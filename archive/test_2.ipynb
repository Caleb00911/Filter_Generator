```python
class Graph:
    def __init__(self):
        self.nodes = {}
        self.edges = []

    def add_node(self, name, **attrs):
        self.nodes[name] = attrs

    def add_edge(self, src, dst, **attrs):
        self.edges.append((src, dst, attrs))

# Example usage
#G = Graph()  # empty graph
#G.add_node('Pole(8)', type='pole', order=8)
#G.add_edge('Pole(8)', 'Out')
```

```python
def filter(n):
    G = Graph()
    G.add_node('Pole', order = 8)
    return G
```

```python
G = filter(8)
G.nodes
order = G.nodes['Pole']['order']
print(order)
```

```python
def combos(n):
    if n == 0:
        return [[]]   # 1 way: empty combination
    if n == 1:
        return [[1]]  # only [1]

    results = []

    # For each way to make n-1, add a 1 at the end
    for c in combos(n-1):
        results.append(c + [1])

    # For each way to make n-2, add a 2 at the end
    for c in combos(n-2):
        results.append(c + [2])

    return results
```

```python
r = combos(G.nodes['Pole']['order'])
print(r)
```

```python
def build_composistions(list_combos, graph):
    graph = graph
    composistions = list_combos
    name = 1
    for i in composistions:
        graph.add_node(f"Composistion{name}", ordering = i)
        graph.add_edge('Pole', f"Composistion{name}")
        name += 1
```

```python
build_composistions(r, G)
```
